```python
df.drop(['stn_code','agency','location_monitoring_station','sampling_date'],axis = 1)
```

```python
df.dropna(subset='date')
```

```python
df['date'] = pd.to_datetime(df['date'],errors = 'coerce')
df.head()
```

```python
df['year']=df.date.dt.year
df.head()
```

```python
# Missing Values
column = ['so2','no2','rspm','spm','pm2_5']
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan , strategy= 'mean')
df[column]=imputer.fit_transform(df[column])
```

```python
df.head()
```

```python
#Data Transformation
df['so2'].transform(func = lambda x : x+2.0)
plt.hist(df['so2'])
```
